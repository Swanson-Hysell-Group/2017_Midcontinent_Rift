# Michipicoten Structural Analysis

Analysis of structural measurements made on Michipicoten Island, 2015. Included in this notebook is a [potentially instructive analysis](#Channel-Lake-Member) on extracting strike/dip measurements from orthogonal orientations (e.g. columnar jointing).

## AP

```python
# import structural measurements
ALL_planes = pd.DataFrame(columns=['strike','dip', 'dip_dir','notes'])
ALL_planes.loc[0] = [75, 15, 0, 'flow top']
ALL_planes.loc[1] = [78, 13, 0, 'flow top']
ALL_planes.loc[2] = [81, 14, 0, 'flow top']
ALL_planes.loc[3] = [70, 15, 0, 'flow top']
ALL_planes.loc[4] = [79, 14, 0, 'flow top']
ALL_planes.loc[5] = [78, 13, 0, 'flow top']
ALL_planes.loc[6] = [81, 14, 0, 'flow top']
ALL_planes.loc[7] = [50, 13, 0, 'pahoehoe']
ALL_planes.loc[8] = [78, 14, 0, 'pahoehoe']
ALL_planes.loc[9] = [59, 11, 0, 'pahoehoe']
ALL_planes.loc[10] = [92, 13, 0, 'pahoehoe']
#ALL_planes.loc[11] = [102, 39, 0, 'flow top']
#ALL_planes.loc[12] = [82, 26, 0, 'flow top']
```

# Agate Point

```python
AP_analysis = pd.DataFrame(columns=["dip","dip_dir","pole_trend", "pole_plunge"])
AP_analysis["dip"] = ALL_planes['dip']
AP_analysis["dip_dir"] = (ALL_planes['strike']+90)%360
AP_analysis['pole_trend'] = (ALL_planes['strike']+270)%360
AP_analysis['pole_plunge'] = 90 - ALL_planes['dip']

AP_analysis
```

```python
AP_poles = ipmag.make_di_block(AP_analysis['pole_trend'],AP_analysis['pole_plunge'])
AP_poles_mean = pmag.fisher_mean(AP_poles)
AP_poles_mean
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='stereonet')

AP_mean_strike = AP_poles_mean['dec'] + 90.0
AP_mean_dip = 90.0 - AP_poles_mean['inc']

ax.plane(AP_analysis['dip_dir']-90,AP_analysis['dip'], 'g-', linewidth=1.5)
ax.pole(AP_analysis['dip_dir']-90,AP_analysis['dip'], 'bo', markersize=8)
ax.plane(AP_mean_strike,AP_mean_dip, 'r-', linewidth=3)
ax.pole(AP_mean_strike,AP_mean_dip, 'r^', markersize=15)
ax.grid()

plt.show()
print AP_mean_strike%360,AP_mean_dip
```
